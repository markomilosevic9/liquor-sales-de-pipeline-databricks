```python
%sql
-- generate the full 2025 calendar as dim_dates
CREATE OR REPLACE TABLE dev.silver.dim_dates AS
WITH dates AS (
    SELECT explode(sequence(
        DATE '2025-01-01',
        DATE '2025-12-31',
        INTERVAL 1 DAY
    )) AS date_day
)
SELECT
    date_day,
    DAYOFWEEK(date_day)                                     AS day_of_week,
    DATE_FORMAT(date_day, 'EEEE')                           AS day_name,
    DAY(date_day)                                           AS day_of_month,
    DAYOFYEAR(date_day)                                     AS day_of_year,
    WEEKOFYEAR(date_day)                                    AS week_of_year,
    MONTH(date_day)                                         AS month_num,
    DATE_FORMAT(date_day, 'MMMM')                           AS month_name,
    QUARTER(date_day)                                       AS quarter,
    YEAR(date_day)                                          AS year,
    CASE WHEN DAYOFWEEK(date_day) IN (1, 7) THEN TRUE
         ELSE FALSE END                                     AS is_weekend,
    CASE WHEN date_day = LAST_DAY(date_day) THEN TRUE
         ELSE FALSE END                                     AS is_month_end
FROM dates;
```

```python
%sql
-- verify
SELECT COUNT(*) AS total_days FROM dev.silver.dim_dates;
-- should return 365

SELECT * FROM dev.silver.dim_dates LIMIT 10;
-- first 10 days of january 2025

SELECT day_name, COUNT(*) AS occurrences
FROM dev.silver.dim_dates
GROUP BY day_name
ORDER BY MIN(day_of_week);
-- should show ~52 of each day
```
