```python
# dim_stores: SCD 2 via APPLY CHANGES INTO
import dlt

dlt.create_streaming_table(
    name="dim_stores",
    comment="SCD type 2 for Stores, tracks name, address, city, zip, county changes"
)

dlt.apply_changes(
    target="dim_stores",
    source="stg_store_changes",
    keys=["store_id"],
    sequence_by="last_seen_date",
    stored_as_scd_type=2,
    track_history_column_list=[
        "store_name", "address", "city", "zip_code", "county"
    ]
)
```

```python
# dim_products: SCD 2
dlt.create_streaming_table(
    name="dim_products",
    comment="SCD Type 2 for Products, tracks description, category code, category name changes"
)

dlt.apply_changes(
    target="dim_products",
    source="stg_product_changes",
    keys=["item_id"],
    sequence_by="last_seen_date",
    stored_as_scd_type=2,
    track_history_column_list=[
        "item_description", "category_code", "category_name"
    ]
)
```
